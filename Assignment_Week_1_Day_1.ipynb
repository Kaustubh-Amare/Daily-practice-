## 6. What is Semi- Supervised Learning and why is it so important?


Semi-supervised learning is a type of machine learning that uses a combination of labeled and unlabeled data to train a model. It is important because it can help to reduce the cost and time required to label large datasets, and can improve the performance of models by leveraging the information contained in both labeled and unlabeled data. Semi-supervised learning can also be used in scenarios where labeled data is scarce or expensive to obtain, making it a valuable technique for a wide range of real-world applications.

## 7. Why we can’t do a classification problem using Regression?

Classification and regression are two different types of machine learning problems that involve predicting different types of outcomes. In classification, the goal is to predict a categorical variable, while in regression, the goal is to predict a continuous variable. Since these two types of problems involve different types of predictions, it is not appropriate to use a regression model to solve a classification problem. Regression models are designed to make continuous predictions, while classification problems require discrete predictions, which can only be achieved using appropriate algorithms such as logistic regression, decision trees.

## 8. How do you treat heteroscedasticity in regression? What causes Heteroscedasticity and how to fix it?

Heteroscedasticity in regression occurs when the variance of the residuals is not constant across all levels of the independent variable(s). This violates the assumption of homoscedasticity, which can lead to biased and inefficient estimates of the regression coefficients. To treat heteroscedasticity, one can use the following techniques:

* Transform the dependent variable: One way to treat heteroscedasticity is to transform the dependent variable using a log or square root transformation.

* Use weighted least squares: Another method is to use weighted least squares regression, where the weights are assigned based on the variance of the residuals.

* Use robust standard errors: Robust standard errors are a method for estimating the standard errors of the regression coefficients, which are robust to heteroscedasticity.

* Use a different regression model: If the above methods do not work, one can use a different regression model that is more appropriate for heteroscedastic data, such as the generalized linear model or the nonlinear regression model.

Heteroscedasticity can be caused by a variety of factors, such as measurement error, outliers, or missing variables. To fix it, one needs to identify the cause and then apply the appropriate treatment method.
